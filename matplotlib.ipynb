```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Krakow Area Survey")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

```python
# Object Oriented
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.15, 0.4, 0.3])
axes1.plot(x,y)
axes2.plot(y,x)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
    current_ax.set_title("Plot")
plt.show()
```
