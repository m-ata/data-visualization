```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.displot(tips["total_bill"], kde=False, bins=100)
plt.show()
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
plt.show()
```

```python
sns.pairplot(tips, hue="sex", palette="coolwarm")
plt.show()
```

```python
sns.rugplot(tips["total_bill"])
plt.show()
```

```python
sns.displot(tips["total_bill"], kde=False)
plt.show()
```

```python
sns.kdeplot(tips["total_bill"])
plt.show()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
plt.show()
```

```python
sns.countplot(x='sex',data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='smoker')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
t = tips.corr(numeric_only=True)
```

```python
sns.heatmap(t, annot=True, cmap="coolwarm")
plt.show()
```

```python
fp = flights.pivot_table(index="month",columns="year",values="passengers",observed=False)
```

```python
sns.heatmap(fp, linecolor="black", linewidth=3)
plt.show()
```

```python
sns.clustermap(fp, standard_scale=1, cmap="coolwarm")
plt.show()
```
